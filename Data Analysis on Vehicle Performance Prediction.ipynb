```python
df[df['horsepower']=='?'].index
df.drop(index= df[df['horsepower']=='?'].index, inplace= True)
```

```python
df['horsepower']=df['horsepower'].astype('int')
```

```python
from sklearn.preprocessing import RobustScaler
```

```python
rs= RobustScaler()
```

```python
from sklearn.linear_model import SGDRegressor
```

```python
sgd= SGDRegressor()
```

```python
df['origin']= df['origin'].astype('category')
df['cylinders']= df['cylinders'].astype('category')
df['year']= df['year'].astype('category')
```

```python
X= df.drop(['Kilometer_per_liter','name'], axis=1)
```

```python
y= df['Kilometer_per_liter']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test= train_test_split(X,y, test_size=0.3)
```

```python
Scaled_train= rs.fit_transform(X_train)
Scaled_test= rs.transform(X_test)
```

```python
sgd.fit(Scaled_train,y_train)
```

```python
y_pred= sgd.predict(Scaled-_test)
```

```python
from sklearn.metrics import mean_absolute_percentage_error,r2_score
```

```python
mean_absolute_percentage_error(y_test,y_pred)
```

```python
r2_score(y_test,y_pred)
```
